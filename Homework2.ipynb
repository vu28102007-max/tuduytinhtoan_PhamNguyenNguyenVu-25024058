Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diệntíchvàchu vi của nó.

Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.

```python
r = float(input("Ban kinh hinh tron: "))
dientich = 3.14 * (r**2)
chuvi = 2 * 3.14 * r
print('dien tich va chu vi cua hinh tron lan luot la: ', dientich, chuvi)
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a, b, c = map(float, input().split())
p = (a + b + c) / 2
dientich = (p * (p - a) * (p - b) * (p - c)) ** 0.5
chuvi = a + b + c
if a + b > c and a + c > b and b + c > a and a > 0 and b > 0 and c > 0:
    if a == b == c:
        print('Tam giac deu')
    elif a == b or a == c or b == c:
        print('Tam giac can')
    elif a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
        print('Tam giac vuong')
    else:
        print('Tam giac thuong')
print('dien tich va chu vi cua tam giac lan luot la: ', dientich, chuvi)
```

Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0) 

```python
a = float(input("Nhap so a: "))
b = float(input("Nhap so b: "))
if a == 0:
    if b == 0:
        print('Phuong trinh co vo so nghiem')
    else:
        print('Phuong trinh vo nghiem')
if a != 0:
    nghiem = -b / a
    print('Phuong trinh co nghiem duy nhat: ', nghiem)
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a = float(input("Nhap so a: "))
b = float(input("Nhap so b: "))
c = float(input("Nhap so c: "))
if a == 0:
    if b == 0:
        if c == 0:
            print('Phuong trinh co vo so nghiem')
        else:
            print('Phuong trinh vo nghiem')
    else:
        nghiem = -c / b
        print('Phuong trinh co nghiem duy nhat: ', nghiem)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print('Phuong trinh vo nghiem')
    elif delta == 0:
        nghiem = -b / (2*a)
        print('Phuong trinh co nghiem kep: ', nghiem)
    else:
        nghiem1 = (-b + delta**0.5) / (2*a)
        nghiem2 = (-b - delta**0.5) / (2*a)
        print('Phuong trinh co hai nghiem phan biet: ', nghiem1, nghiem2)
```

Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bànphím).

```python
a, b, c, d = map(float, input().split())
max = a
if b > max:
    max = b
    if c > max:
        max = c
        if d > max:
            max = d
print("Gia tri lon nhat la:", max)
```

Bài 8: Giải hệ phương trình tuyến tính:
- ax + by = m
- cx + dy = n

```python
a, b, m, c, d, n = map(float, input().split())
if a/c == b/d == m/n:
    print('phương trình có vô số nghiệm')
if a/c == b/d != m/n:
    print('phương trình vô nghiệm')
if a/c != b/d:
    y = (m * c - a * n) / (b * c - a * d)
    x = (n - d * y) / c
    print('phương trình có nghiệm duy nhất: x =', x, 'y =', y)
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây.

```python
x = int(input())
h = x // 3600
m = (x - h * 3600) // 60
s = x - h * 3600 - m * 60
print('Thoi gian sau khi chuyen doi la: ', h, 'gio', m, 'phut', s, 'giay')
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x0 = float(input())
y0 = float(input())
r = float(input())
x = float(input("Nhập xA: "))
y = float(input("Nhập yA: "))
if (x - x0)**2 + (y - y0)**2 <= r**2:
    print('Điểm A thuộc đường tròn') 
else:
    print('Điểm A không thuộc đường tròn')
```
