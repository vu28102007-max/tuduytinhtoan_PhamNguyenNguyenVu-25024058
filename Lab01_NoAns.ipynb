### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
def count_k(n):
    count = 0
    for k in range(1, n):
        if k > (n - 1) / 2:
            count += 1
    return count
print(count_k(9))
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
import math

def calculate_triangle_area(a, b, c):
    if a <= 0 or b <= 0 or c <= 0 or (a + b <= c) or (a + c <= b) or (b + c <= a):
        raise ValueError("Ba độ dài đã cho không thể tạo thành một tam giác hợp lệ.")
    p = (a + b + c) / 2
    area = math.sqrt(p * (p - a) * (p - b) * (p - c))
    return area
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
def convert_to_lowercase(char):
    if not isinstance(char, str) or len(char) != 1:
        raise ValueError("Đầu vào phải là một ký tự duy nhất.")
    return char.lower()
```

### Problem 10
#### Write a function that, given a lowercase letter, returns a string containing that letter repeated 10 times.

```python
def repeat_lowercase_letter(char):
    if not isinstance(char, str) or len(char) != 1 or not char.islower():
        raise ValueError("Input must be a single lowercase letter.")
        return char * 10
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
def generate_triangle(n):
    lines = ['*' * i for i in range(1, n + 1)]
    return '\n'.join(lines)
print(generate_triangle(10))
```
