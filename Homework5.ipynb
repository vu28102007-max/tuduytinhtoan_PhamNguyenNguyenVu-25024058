Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x: int) -> bool:
    if x < 0:
        return False
    str_x = str(x)
    num_digits = len(str_x)
    total_sum = sum(int(digit) ** num_digits for digit in str_x)
    return total_sum == x
if __name__ == "__main__":
    n = 153
    if is_armstrong(n):
        print(f"{n} là số Armstrong.")
    else:
        print(f"{n} không phải là số Armstrong.")
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_perfect_square(x: int) -> bool:
    if x < 0:
        return False
    if x == 0 or x == 1:
        return True   
    left = 1
    right = x // 2
    while left <= right:
        mid = left + (right - left) // 2
        square = mid * mid    
        if square == x:
            return True
        elif square < x:
            left = mid + 1
        else:
            right = mid - 1
    return False
if __name__ == "__main__":
    n = 16
    if is_perfect_square(n):
        print(f"{n} là số chính phương.")
    else:
        print(f"{n} không phải là số chính phương.")
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_palindrome(x: int) -> bool:
    if x < 0 or (x % 10 == 0 and x != 0):
        return False
    reversed_half = 0
    while x > reversed_half:
        reversed_half = reversed_half * 10 + x % 10
        x //= 10
    return x == reversed_half or x == reversed_half // 10

if __name__ == "__main__":
    test_cases = [121, -121, 10, 1221, 12321]
    for n in test_cases:
        if is_palindrome(n):
            print(f"{n} là số đối xứng.")
        else:
            print(f"{n} không phải là số đối xứng.")
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_magic_number(x: int) -> bool:
    if x <= 0:
        return False
    sum_digits = x
    while sum_digits > 9:
        temp = sum_digits
        sum_digits = 0
        while temp > 0:
            sum_digits += temp % 10
            temp //= 10
    return sum_digits == 1
if __name__ == "__main__":
    test_cases = [1, 19, 50113, 22]
    for n in test_cases:
        if is_magic_number(n):
            print(f"{n} là số kỳ diệu.")
        else:
            print(f"{n} không phải số kỳ diệu.")
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'. 

```python
def detect_number_type(s: str) -> str:
    s = str(s).strip()
    if not s:
        return 'neither'
    if s[0] in ('-', '+'):
        body = s[1:]
    else:
        body = s
    if not body:
        return 'neither'
    if body.isdigit():
        return 'int'
    if body.count('.') == 1:
        left, right = body.split('.')
        if left == "" and right == "":
            return 'neither'
        is_left_valid = (left == "" or left.isdigit())
        is_right_valid = (right == "" or right.isdigit())
        if is_left_valid and is_right_valid:
            return 'float'
    return 'neither'
test_cases = ["123", "-45", "3.14", "-.99", "5.", ".", "-", "abc", "12.34.56"]
for test in test_cases:
    print(f"'{test}': {detect_number_type(test)}")
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.


```python
def quadratic_roots_count(a: float, b: float, c: float):
    if a == 0:
        if b != 0:
            return 1
        else:
            if c == 0:
                return float('inf')
            else:
                return 0
    delta = b**2 - 4 * a * c
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1, y1, x2, y2, x3, y3) -> bool:
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)


```python
def count_nonzero_digits_math(numbers: list) -> int:
    count = 0
    for num in numbers:
        temp = abs(num)
        if temp == 0:
            continue
        while temp > 0:
            digit = temp % 10
            if digit != 0:
                count += 1
            temp //= 10
    return count
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0

```python
def sum_1_to_n_math(n: int) -> int:
    if n <= 0:
        return 0
    total_sum = 0
    for x in range(1, n + 1):
        temp = x
        seen_digits = set()
        is_unique = True
        while temp > 0:
            digit = temp % 10
            if digit in seen_digits:
                is_unique = False
                break
            seen_digits.add(digit)
            temp //= 10
        if is_unique:
            total_sum += x
    return total_sum
```
