W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a, b = map(int, input().split())
min_value = a if a < b else b
print(f"Các ước chung của {a} và {b} là:")
for i in range(1, min_value + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=' ')
```

W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input())
tamthoi = n
while tamthoi % 2 == 0:
    tamthoi //= 2
if tamthoi == 1:
    print(f"{n} là số nguyên dương có dạng 2^k")
else:
    print(f"{n} không phải là số nguyên dương có dạng 2^k")
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input())
tamthoi = n
Sum = 0
while tamthoi > 0:
    chu_so = tamthoi % 10
    Sum += chu_so
    tamthoi //= 10
print(f"Tổng các chữ số của {n} là: {Sum}")
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
tamthoi = n
sochan = 0
sole = 0
while tamthoi > 0:
    chuso = tamthoi % 10
    if chuso % 2 == 0:
        sochan += 1
    else:
        sole += 1
    tamthoi //= 10
print(f"Số lượng chữ số chẵn và số lẻ lần lượt là: {sochan} và {sole}")
```

Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
tong = 0
k = 0
while tong + (k + 1) < n:
    k += 1
    tong += k
print("Số k sao cho S(k) lớn nhất nhưng nhỏ hơn n là:", k)
```

W4A24 Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input())
s = 0
n = 1
while s <= A:
    n += 1
    s += 1/n
print("Số n sao cho tổng lớn hơn A là:", n)
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
x = int(input())
if x == -1:
    print("Bạn chưa nhập số nào")
else:
    lon_nhat = x
    nho_nhat = x
while True:
    x = int(input())
    if x == -1:
        print("Bạn đã nhập -1, chương trình kết thúc.")
        break
    if x > lon_nhat:
        lon_nhat = x
    if x < nho_nhat:
        nho_nhat = x
print(f"Số lớn nhất là: {lon_nhat}")
print(f"Số nhỏ nhất là: {nho_nhat}")
```

W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
A = int(input("Nhập vào số nguyên dương A: "))
n1 = 1
n2 = 1
if A == 1:
    print(f"Số {A} là số Fibonacci.")
else:
    kqua = 1
    while True:
        n3 = n1 + n2
        if n3 > A:
            break
        kqua = n3
        n1 = n2
        n2 = n3
    print(f"Số fibonacci lớn nhất nhỏ hơn hoặc bằng {A} là: {kqua}")
```

```python
s = input("Nhập vào một chuỗi: ")
length = 0
for char in s:
    length += 1
so_tu = 0
for i in range(length):
    if s[i] != " ":
        if i == 0 or s[i-1] == " ":
            so_tu += 1
print("Số từ trong chuỗi là:", so_tu)
```

W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa.

```python
s = input()
kq = ""
for char in s:
    if char.isupper():
        kq += char.lower()
    else:
        kq += char.upper()
print("Chuỗi sau khi đổi chữ hoa thành chữ thường và ngược lại là:", kq)
```

W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
s = input()
hoa = 0
thuong = 0
so = 0
for char in s:
    if char.isupper():
        hoa += 1
    elif char.islower():
        thuong += 1
    elif char.isdigit():
        so += 1
print(f"Số ký tự in hoa: {hoa}")
print(f"Số ký tự in thường: {thuong}")
print(f"Số ký tự chữ số: {so}")
```
