Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập vào chiều cao của tam giác: "))
for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - 1) + "*")
    elif i == h:
        print((2 * h -1) * "*")
    else:
        print(" " * (h - i) + "*" + " " * (2 * i - 3) + "*")
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. 
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
a = int(input("Nhập vào số nguyên dương a: "))
tong_cac_uoc  = 0
for i in range(1, a + 1):
    if a % i == 0:
        tong_cac_uoc += i
if tong_cac_uoc == a:
    print(f"Số {a} là số hoàn hảo.")
else:
    print(f"Số {a} không phải là số hoàn hảo.")
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) 
lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhap vao so nguyen duong n: "))
k = 0
S = 0
while S + k < n:
    k += 1
    S += k
print(f"Số nguyên dương k lớn nhất sao cho S(k) < n là: {k}")
print(f"Tổng S(k) = {S}")
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
x = int(input())
if x == -1:
    print("Bạn chưa nhập số nào")
else:
    lon_nhat = x
    nho_nhat = x
while True:
    x = int(input())
    if x == -1:
        print("Bạn đã nhập -1, chương trình kết thúc.")
        break
    if x > lon_nhat:
        lon_nhat = x
    if x < nho_nhat:
        nho_nhat = x
print(f"Số lớn nhất là: {lon_nhat}")
print(f"Số nhỏ nhất là: {nho_nhat}")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input("Nhập vào số nguyên dương A: "))
n1 = 1
n2 = 1
if A == 1:
    print(f"Số {A} là số Fibonacci.")
else:
    kqua = 1
    while True:
        n3 = n1 + n2
        if n3 > A:
            break
        kqua = n3
        n1 = n2
        n2 = n3
    print(f"Số fibonacci lớn nhất nhỏ hơn hoặc bằng {A} là: {kqua}")
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input("Nhap vao so nguyen duong n: "))
tam_thoi = n
demsonguyento = 0
chusonguyento = [2, 3, 5, 7]
while tam_thoi > 0:
        chu_so = tam_thoi % 10
        if chu_so in chusonguyento:
            demsonguyento += 1
        tam_thoi //= 10
print(f"Số {n} có {demsonguyento} chữ số là số nguyên tố.")
```
